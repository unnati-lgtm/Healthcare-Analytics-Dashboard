```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("../dataset/healthcare_dataset.csv")

engine = create_engine(
    "mysql+pymysql://root:1234@localhost:3306/healthcare_analytics"
)

df.to_sql("healthcare", con=engine, if_exists="replace", index=False)

print("Data Imported Successfully!")
```

```python
import pandas as pd

mental = pd.read_csv("../dataset/Mental Health Dataset.csv")

mental.head()
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:1234@localhost:3306/healthcare_analytics"
)
```

```python
mental.to_sql(
    "mental_health",
    con=engine,
    if_exists="replace",
    index=False
)

print("Mental Health Data Imported Successfully!")
```
